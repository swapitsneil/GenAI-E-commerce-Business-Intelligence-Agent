```python
import pandas as pd

df = pd.read_csv('data.csv', encoding='latin-1')

# head and tail
print("Head of the dataset:")
print(df.head())
print("\nTail of the dataset:")
print(df.tail())
print("\nShape of the dataset:", df.shape)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df['Profit'] = df['Quantity'] * df['UnitPrice']
total_profit = df['Profit'].sum()
print(f"Total Profit: ${total_profit:.2f}")
df.head()
```

```python
df['InvoiceDateTime'] = pd.to_datetime(df['InvoiceDate'], format='%m/%d/%Y %H:%M')
df['Invoice_Month'] = df['InvoiceDateTime'].dt.month_name()
df['Invoice_Date'] = df['InvoiceDateTime'].dt.date
df['Invoice_Time'] = df['InvoiceDateTime'].dt.time
```

```python
df.head()
```

dropping the null values from description 1454 and customerid 135080

```python
df = df.dropna(subset=['Description', 'CustomerID'])
df.shape
```

```python
# countries where products have been sold
df['Country'].unique()
```

```python
# Group by country and sum the quantity
country_sales = df.groupby('Country')['Quantity'].sum().sort_values(ascending=False)

print("Countries by Total Quantity Sold (Descending Order):")
print(country_sales)
```

```python
# Count number of transactions per country
country_transactions = df.groupby('Country')['InvoiceNo'].count().sort_values(ascending=False)

print("Countries by Number of Transactions (Descending Order):")
print(country_transactions)
```

```python
# Group by country and sum the profit
country_profit = df.groupby('Country')['Profit'].sum().sort_values(ascending=False)

print("Countries by Total Profit (Descending Order):")
print(country_profit)
```

```python
# Comprehensive country analysis
country_analysis = df.groupby('Country').agg({
    'InvoiceNo': 'count',           # no. of transactions
    'Quantity': 'sum',              # total units sold
    'Profit': 'sum',               # total revenue/profit
    'CustomerID': 'nunique'        # unique customers
}).sort_values('Profit', ascending=False)

country_analysis = country_analysis.rename(columns={
    'InvoiceNo': 'Total_Transactions',
    'Quantity': 'Total_Units_Sold',
    'Profit': 'Total_Profit',
    'CustomerID': 'Unique_Customers'
})

print("Complete Country Sales Analysis (Sorted by Profit):")
print(country_analysis)
```

```python
# top countries by profit with percentage

country_profit = df.groupby('Country')['Profit'].sum().sort_values(ascending=False)

total_profit = country_profit.sum()
country_percentage = (country_profit / total_profit * 100).round(2)

country_summary = pd.DataFrame({
    'Total_Profit': country_profit,
    'Percentage_Contribution': country_percentage
})

print("Country Sales with Percentage Contribution:")
print(country_summary)
```

```python
# Get top 10 countries
top_countries = country_profit.head(10)

print("Top 10 Countries by Sales:")
print(top_countries)
```

```python
plt.figure(figsize=(12, 6))
top_countries.plot(kind='bar', color='purple')
plt.title('Top 10 Countries by Total Profit')
plt.xlabel('Country')
plt.ylabel('Total Profit')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# countries by total profit in descending order
print(df.groupby('Country')['Profit'].sum().sort_values(ascending=False))
```

```python
# top 5 by profit
print("Top 5 Products by Profit:")
print(df.groupby(['StockCode', 'Description'])['Profit'].sum().nlargest(5))
```

```python
top_5_profit = df.groupby(['StockCode', 'Description'])['Profit'].sum().nlargest(5)

plt.figure(figsize=(10, 8))
top_5_profit.plot(kind='pie', autopct='%1.1f%%',
                  labels=[desc[:20] + "..." for desc in top_5_profit.index.get_level_values(1)])
plt.title('Top 5 Products by Profit')
plt.show()
```

```python
# UK top products
uk_top = df[df['Country']=='United Kingdom'].groupby('Description')['Profit'].sum().nlargest(5)

print("UK Top 5 Products:")
print(uk_top)

uk_top.plot(kind='barh', title='Top 5 Products in UK')
plt.show()
```

```python
# Convert to datetimh
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df['Month'] = df['InvoiceDate'].dt.to_period('M')

# Monthly sales
monthly_sales = df.groupby('Month')['Profit'].sum()

print("Monthly Sales Trend:")
for month, sales in monthly_sales.items():
    print(f"{month}: ${sales:,.2f}")

monthly_sales.plot(kind='line', marker='o', figsize=(12, 6))
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Revenue ()')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Most selling months by total profit

monthly_sales = df.groupby('Invoice_Month')['Profit'].sum().sort_values(ascending=False)

print("Months by Total Sales (Descending Order):")
for month, sales in monthly_sales.items():
    print(f"{month}: ${sales:,.2f}")
```

```python
# Bar chart
plt.figure(figsize=(10, 6))
monthly_sales.plot(kind='bar', color='orange')
plt.title('Total Sales by Month')
plt.xlabel('Month')
plt.ylabel('Total Profit ()')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
